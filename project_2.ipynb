```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
null_rows = df[df.isnull().any(axis=1)]
print(null_rows)
```

```python
df.info()
```

```python
ax=sns.countplot(x='species',data=df,hue='sex',palette='Accent')
plt.title('Species_count_on_the_basis_of_Sex')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
ax=sns.countplot(x='species',data=df,hue='island',hue_order=['Torgersen',  'Dream','Biscoe'])
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Counting_of_species_on_the_basis_of_islands",fontsize=10)
plt.show()
```
